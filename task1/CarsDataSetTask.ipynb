```python

# updated the column names Like(To Start with Upper Litter)
aa={}
for i in list(data.columns):
  aa[i] = i.capitalize()
data.rename(columns={'Web-scraper-order': 'Web-scraper-order', 'Car model': 'Car model', 'Month/year': 'Month/year', 'Average price': 'Average price', 'Minimum price': 'Minimum price', 'Maximum price': 'Maximum price'}
, inplace=True)
data.head()
```

```python
# Split Car Full Data into Brand, Model, And Year of The Model
data[['Brand', 'Model', 'Year']] = data['Car model'].str.split(' ', 2, expand=True)
# data[['Model', 'Year']] = data['Model'].str.split(' ', 2, expand=True)
data
```

```python
# @title Default title text
# Convert Average price, Minimum price, Maximum price to Numaarical Values
data['Average price'] = data['Average price'].str.replace('EGP', '')
data['Average price'] = data['Average price'].str.replace(',', '')
data['Average price'] = data['Average price'].astype(float)
data['Average price']
```

```python
data['Minimum price'] = data['Minimum price'].str.replace('EGP', '')
data['Minimum price'] = data['Minimum price'].str.replace(',', '')
data['Minimum price'] = data['Minimum price'].astype(float)
data['Minimum price']
```

```python
data['Maximum price'] = data['Maximum price'].str.replace('EGP', '')
data['Maximum price'] = data['Maximum price'].str.replace(',', '')
data['Maximum price'] = data['Maximum price'].astype(float)
data['Maximum price']
```

```python
# Split Month/Year Column to Year and Month
data[['Year', 'Month']] = data['Month/year'].str.split('-', 1, expand=True)
data
```

```python
# let's start with (Average price ,Minimum price ,Maximum price ) , see the number of nulls in them , replace these nulls with the ( mean , median and mode) of these columns
#and in the end check that there are not nulls in data
null_counts = data[['Average price', 'Minimum price', 'Maximum price']].isnull().sum()
print("Number of null values in each column:")
print(null_counts)

mean_values = data[['Average price', 'Minimum price', 'Maximum price']].mean()
median_values = data[['Average price', 'Minimum price', 'Maximum price']].median()
mode_values = data[['Average price', 'Minimum price', 'Maximum price']].mode().iloc[0]  # mode() returns a DataFrame

data['Average price'].fillna(mean_values['Average price'], inplace=True)
data['Minimum price'].fillna(median_values['Minimum price'], inplace=True)
data['Maximum price'].fillna(mode_values['Maximum price'], inplace=True)


null_counts = data[['Average price', 'Minimum price', 'Maximum price']].isnull().sum()
print("Number of null values in each column:")
print(null_counts)
```
